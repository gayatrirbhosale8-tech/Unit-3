```python
from nltk.tokenize import word_tokenize
text = "Natural language processing is interesting"
tokens = word_tokenize(text)
print(tokens)
```

```python
#Sentence Tokenization
from nltk.tokenize import sent_tokenize
text = "Natural language processing is interesting. It is used in chatbots."
sentences = sent_tokenize(text)
print(sentences)

```

```python
#Stop Word Removal
#Stop words such as is,the,and,in are removed

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

```python
#Stemming
#Stemming reduces words to their root form

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["running", "rans", "drawing"]

for w in words:
  print(stemmer.stem(w))
```

```python
#Lemmantization
#Lemmantization converts words to their dictionary form
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("running"))
print(lemmatizer.lemmatize("drawing"))
```

```python
#Part-of-speech (POS) Tagging
#POS Tagging identifies grammatical roles of words

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

```

```python
from nltk import pos_tag
words = word_tokenize("NLP echniques On Text Data In Python Is Very Intresting")
pos_tags = pos_tag(words)
print(pos_tags)
```

```python
#Word Frequency Count
from nltk.probability import FreqDist
word = word_tokenize(text)
fd = FreqDist(word)
print(fd.most_common())
```
