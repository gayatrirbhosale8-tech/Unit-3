```python
import pandas as pd
data={
    'Order_ID':['01','02','03','04','05','06','07','08','09','10'],
    'Category':['Furniture','Office Supplies','Furniture','Technology','Technology','Furniture','Office Supplies','Furniture','Technology','Technology'],
    'Payment_Method':['Cash','UPI','Cash','UPI','UPI','Cash','UPI','Cash','UPI','Cash'],
    'Delivery_type':['Standard Class','Express Class','Standard Class','Express Class','Standard Class','Standard Class','Express Class','Standard Class','Express Class','Express Class'],
    'Customer_type':['New','Returning','Returning','New','Returning','New','Returning','New','Returning','New'],

}

df=pd.DataFrame(data)
print(df)
```

```python
#1. Frequency Count
#Count How many orders in each category

print(df['Category'].value_counts())
```

```python
#2. Unique Values
#Display all unique categories.
print(df['Category'].unique())
```

```python
#3. Number of unique values
print(df['Category'].nunique())
```

```python
#Frequency of payment methods
#1. Frequency Count
#count How Many Orders In Each Payment_Method.
print(df['Payment_Method'].value_counts())
```

```python
#2. Unique Values
#Display all unique Payment_Method.
print(df['Payment_Method'].unique())
```

```python
#3. Number of unique values
print(df['Payment_Method'].nunique())
```

```python
#5.Cross Tabulation
#Category Vs Payment Method
print(pd.crosstab(df['Category'],df['Payment_Method']))
```

```python
#6. Cross Tabulation
#Delivery Vs Customer Type
print(pd.crosstab(df['Delivery_type'],df['Customer_type']))
```

```python
#7. Percentage Distribution
print(df['Category'].value_counts(normalize=True)*100)
```

```python
#8. Filtering Data
#Example: Order From Technology Category
Technology_orders=df[df['Category']=='Technology']
print(Technology_orders)
```

```python
#Example: Order From  Office Supplies Category
Office_Supplies=df[df['Category']=='Office Supplies']
print(Office_Supplies)
```

```python
#9.Grouping
#Orders by category and payment method
print(df.groupby('Category')['Payment_Method'].value_counts())
```

```python
#10. Sorting Data
sorted_df = df.sort_values(by='Category')
print(sorted_df)
```
