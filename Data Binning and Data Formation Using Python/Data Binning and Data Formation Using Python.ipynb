```python
import pandas as pd
import numpy as np

data ={
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units_Sold':[25,60,35,80,50,15]
}
df = pd.DataFrame(data)
print("Dataset")
print(df)
```

```python
#Data Binnning
bins1 = [0,10000,30000,60000]
labels1 = ['Low','Medium','High']
df['Price_Category'] = pd.cut(df['Price'],bins=bins1,labels=labels1)
print(df)
```

```python
#Binning Units Sold
bins2 = [0,30,60,100]
labels2 = ['Low sales','Moderate Sales','High Sales']
df['Sales Category'] = pd.cut(df['Units_Sold'],bins=bins2,labels=labels2)
print(df)
```

```python
#Data Formating
#Convert Data Type
print(df.dtypes)
df['Units_Sold']=df['Units_Sold'].astype(float)
print(df.dtypes)
```

```python
#Convert Text to Uppercase
df['Product']=df['Product'].str.upper()
print(df)
```

```python
#Rounding Values
df['Price']=df['Price'].round(2)
print(df)
```

```python
 #Additional Operations
 #Sort Data
 sorted_df = df.sort_values(by='Price',ascending=False)
 print(sorted_df)
```

```python
#Display Uniques Values
print(df['Price_Category'].unique())
```

```python
#DataSet: Food Delivery
import pandas as pd
import numpy as np
data={
    'Order_ID':['01','02','03','04','05','06','07','08'],
    'Order_Value':[250,450,120,600,300,150,500,220],
    'Delivery_Time':[30,45,25,60,35,20,50,28],
    'Distance_Km':[5,8,3,10,6,2,9,4]
}
df=pd.DataFrame(data)
print(df)
```

```python
#Binning Order Value
bins1 = [0,200,400,700]
labels1 = ['Low Order','Medium Order','High Order']
df['Order Category'] = pd.cut(df['Order_Value'],bins=bins1,labels=labels1)
df
```

```python
#Binning Delivery Time
bins2 = [0,30,45,70]
lables2=['Fast','Normal','Slow']
df['Delivery_Speed'] = pd.cut(df['Delivery_Time'],bins=bins2,labels=lables2)
df
```

```python
#Data Formating Examples
#Convert Distance To Float
df['Distance_Km']=df['Distance_Km'].astype(float)
df.dtypes
df
```

```python
#Convert Order_ID to Uppercase
df['Order_ID']=df['Order_ID'].str.upper()
df
```

```python
#Check Data Types
print(df.dtypes)
```

```python
#Unique Categories
print(df['Order Category'].unique())
df
```

```python
#Count Categories
print(df['Order Category'].value_counts())
df
```
