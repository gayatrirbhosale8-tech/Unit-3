```python
#Dataset
import pandas as pd
import pandas as pd

data= {
    'Product':['Laptop','Phone','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units_Sold':[100,200,150,50,120,80],
    'Discounts':[10,15,12,20,18,8]
}
df=pd.DataFrame(data)
print("Originl Datset")
print(df)
```

```python
#1. Min-Max Normalization fo price
df['Price_MinMax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())
print("\nMin-Max Normalization")
print(df[['Product','Price','Price_MinMax']])
```

```python
#Min-Max Normalization for Units Sold
df['Units_Sold_MinMax'] = (df['Units_Sold'] - df['Units_Sold'].min()) / (df['Units_Sold'].max() - df['Units_Sold'].min())
print("\nMin-Max Normalization")
print(df[['Product','Units_Sold','Units_Sold_MinMax']])
```

```python
#Min-Max Normalization For Discounts
df['Discounts_MinMax'] = (df['Discounts'] - df['Discounts'].min()) / (df['Discounts'].max() - df['Discounts'].min())
print("\nMin-Max Normalization")
print(df[['Product','Discounts','Discounts_MinMax']])
```

```python
#2. Z-score Noralization
df['Units_Zscore'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()
print("\nZ-score Normalized Units Sold")
print(df[['Product','Units_Sold','Units_Zscore']])
```

```python
#3. Decimal Scaling for price
df['Price_Decimal'] = df['Price'] / 100000
print("\nDecimal Scaled Price")
print(df[['Product','Price','Price_Decimal']])
```

```python
#Decimal Scaling for discounts
df['Discounts_Decimal'] = df['Discounts'] / 100
print("\nDecimal Scaled Discounts")
print(df[['Product','Discounts','Discounts_Decimal']])
```

```python
 #Nomalize Multiple Columns
cols = ['Price', 'Units_Sold', 'Discounts']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nNoralized dataset")
print(df_norm)
```

#Part-B Turning categorical variables into quantitative variables in Python

```python
import pandas as pd
import numpy as np
data = {
    'Order_ID': [101, 102, 103, 104, 105,106],
    'Customer_Gender': ['Male', 'Female', 'Female', 'Male', 'Female','Male'],
    'Product_Category': ['Electronics', 'Clothing', 'Home', 'Electronics','Beauty','Clothing'],
    'Payment_Method': ['UPI', 'Credit Card', 'Debit Card', 'COD', 'UPI','Debit Card'],
    'City':['Pune','Mumbai', 'Delhi', 'Bangalore', 'Pune', 'Hydrabad'],
    'Order_Value':[4500,2300,3200,5200,1800,4100]
}

df = pd.DataFrame(data)
print("Original Dataset")
print(df)
```

```python
#Method 1: Lable Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Gender_Lable'] = label_encoder.fit_transform(df['Customer_Gender'])
print(df)
```

```python
#Method 2: One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Payment_Method'])
print(df_encoded)
```

```python
 #Method 3: Encoding Multiple Columns
df_multi = pd.get_dummies(df, columns=['Product_Category','City'])
print(df_multi)
```

```python
#Method 4: Dummy Encoding (Drop First)
df_dummy = pd.get_dummies(df, columns=['Payment_Method'], drop_first=True)
print(df_dummy)
```
