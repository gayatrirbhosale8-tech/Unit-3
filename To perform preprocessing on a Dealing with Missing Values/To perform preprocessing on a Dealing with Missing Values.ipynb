```python
# Preprocessing of Dataset-1
import pandas as pd
data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha","Rohan","Meera","Karan","Anil"],
    "Age": [20, "-", 19, 21, "22"],
    "Marks": [85,90,"-",88,76],
    "Department": ["CSE","ENTC","CSE","IT","CSE"],
    "Admission_Date":["01-06-2023","03-06-2023","05-06-2023","07-06-2023","10/06/2023"]
}
df = pd.DataFrame(data)
print(df)

# Save as CSV
df.to_csv("student_data.csv", index=False)

```

```python
#Step 2: Replace '-' wiith NaN To Mark missing values
df.replace("-",np.nan,inplace=True)
df
```

```python
#Convert Columns to numeric where needed
df["Age"]=pd.to_numeric(df["Age"], errors='coerce')
df["Marks"]=pd.to_numeric(df["Marks"],errors='coerce')
df

''' When converting data (like strings + datetime),
if some values are invalid or cannot
```

```python
#Step 3: Check missing values
print("\nMissing Values:")
print(df.isna().sum())
```

```python
#Fill Missing Age With Mean
df["Age"].fillna(df["Age"].mean(),inplace=True)
df
```

```python
#Fill Missing Marks With Median
df["Marks"].fillna(df["Marks"].median(),inplace=True)
df
```

```python
#Step 5: Fix inconsistent text formating
df["Department"]=df["Department"].str.upper()
df
```

```python
#Step 6: Convert Admission_Date to standard datetime format
df["Admission_Date"]=pd.to_datetime(df["Admission_Date"],errors='coerce')
df
```

```python
#Step 7:Save cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)

print("ncleaned CSV file saved_student_data")
```

# Preprocessing of Dataset-2
## Checking For Missing Values


```python
#Working with missing values On DataSet-2
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Cars93.csv')
df
```

```python
df.isna().sum() #Attributes are displayed
```

```python
#Implementing the missing Value
#1. Luggage Room
df['Luggage.room']=df['Luggage.room'].fillna(value=df['Luggage.room'].mean())
df.isna().sum()
```

```python
#2. Rear.seat.room
df['Rear.seat.room']=df['Rear.seat.room'].fillna(value=df['Rear.seat.room'].mean())
df.isna().sum()
```

```python
#3 Airbags
df['AirBags']=df['AirBags'].fillna(value=df['AirBags'].mode().values[0])
df.isna().sum()
```

```python
#Save Cleaned DataSet
df.to_csv('updated_cars93.csv', index=False)
```
